この分析では、USGSの地震データをPythonで取得し、1日ごとの地震発生数を折れ線グラフにして可視化しました。取得したデータを日付ごとに整理し、pandasで集計、matplotlibでグラフ化することで、地震が多く発生した日や少なかった日を一目で確認できました。実際にコードを書きながら、APIを使ってデータを取得する方法や、リストやDataFrameの扱い方、グラフの軸やタイトルを整える方法など、Pythonの基本的な分析手順を学ぶことができました。また、日本語フォント設定など、見やすいグラフを作る工夫の大切さも感じました。データを可視化することで、数値だけでは分かりにくい傾向を直感的に理解できる事を実感しました。
